```python
# trazendo a biblioteca capaz de realizar as requisições.
import requests
import random
# variável "url" armazena o endereço com as info que queremos buscar
url = 'https://gist.githubusercontent.com/FabioAlucard/d052f4900aa35ab01b0af75035b606d4/raw/a7ae7b9e27a1bcaf1a4318b502d6a362ad671d08/words.json'
#faço a reuisição e amazeno em uma variável chamada resposta
resposta = requests.get(url)
#transforma a resposta em um JSON
data = resposta.json()
print(len(data))
```

```python
#type(data) - o comando "type" mostra o tipo de dados
#vendo como acessar cada uma das informações em um lista (índice começa em 0)
data[0]
```

```python

#várial chamada valor_secreto que amazena uma técnologia aleatória da lista
valor_secreto = random.choice(data)
#variável para armazenar apenas a palavra
palavra_secreta = valor_secreto['palavra']
#variável para armazenar apenas a dica
dica = valor_secreto['dica']
#print mostrou na tela quantas letras a palavra_secreta possui e a dica
#colchetes indica que o cod deve ser executado com valor diferente da sting
#o "f"serve para juntar palavras e variáveis
print(f'A palavra secreta tem {len(palavra_secreta)} letras')
print(f'Dica: {dica}')
#receber um chute ou palpite da tecnologia
chute = input('O que vc acha que é: ')
if chute == palavra_secreta:
  print('Acertou!')
else:
  print(f'Errou.. a palavra secreta era {palavra_secreta}')
```
